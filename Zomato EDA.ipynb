```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on = "Country Code", how = "left")
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
Country_names = final_df["Country"].value_counts().index
```

```python
## Pie Chart - Top 3 countries that uses zomato

final_df.groupby("Country").size().nlargest(3).plot(kind = "pie",
                                                   autopct = "%.2f")
```

Observation: Zomato is used in India after that USA and then UK

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns = {0:"Rating Count"})
```

```python
ratings
```

## Observation :

1. When the rating is between 4.5 to 4.9 -------> Excellent
2. When the rating is between 4.0 to 4.4 -------> Very Good
3. When the rating is between 3.5 to 3.9 -------> Good
4. When the rating is between 2.5 to 3.4 -------> Average
5. When the rating is between 0.0 to 2.4 -------> Poor

```python
plt.style.use("Solarize_Light2")
sns.barplot(x = "Aggregate rating", y = "Rating Count", data= ratings)
```

```python
plt.style.use('dark_background')

sns.barplot(x = "Aggregate rating", y = "Rating Count",
            data = ratings,
            hue = "Rating text", 
            palette=["white","red","orange","yellow","green","green"])
```

```python
plt.style.use('dark_background')
df.groupby("Aggregate rating").size().plot(kind = "bar")

```

## Observation:
1. Not rated count is very high
2. Maximum number of rating is between 2.5 to 3.4 ------> Average

## Create a pie chart for city distribution 

```python
final_df.groupby("City").size().nlargest(5).plot(kind = "pie",
                                                autopct = "%.2f")
```

```python
city_counts = final_df.City.value_counts().values
city_labels = final_df.City.value_counts().index

plt.pie(city_counts[:5],labels = city_labels[:5],
       autopct = "%.2f")
```
