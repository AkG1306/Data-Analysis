```python
import pandas as pd
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_excel("Data_Train.xlsx")
df_train
```

```python
df_test = pd.read_excel("Test_set.xlsx")
df_test
```

```python
df = pd.concat([df_train,df_test])
```

```python
df
```

```python
df.describe()
```

```python
## Feature engineering
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
```

```python
df
```

```python
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
```

```python
df.drop("Date_of_Journey", axis = 1)
```

```python
df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x: x.split(" ")[0])
```

```python
df
```

```python
df["Arrival_hour"]=df["Arrival_Time"].apply(lambda x: x.split(":")[0])
df["Arrival_minutes"] = df["Arrival_Time"].apply(lambda x:x.split(":")[1])
```

```python
df["Dep_hour"] = df["Dep_Time"].apply(lambda x:x.split(":")[0])
df["Dep_min"] = df["Dep_Time"].apply(lambda x:x.split(":")[1])
```

```python
df["Arrival_hour"]=df["Arrival_hour"].astype(int)
df["Arrival_minutes"]=df["Arrival_minutes"].astype(int)
df["Dep_hour"] = df["Dep_hour"].astype(int)
df["Dep_min"] = df["Dep_min"].astype(int)
```

```python
df.drop(["Arrival_Time","Dep_Time"],axis = 1, inplace = True)
```

```python
df
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,'4 stops':4})
```

```python
df["Total_Stops"].fillna(4, inplace = True)
```

```python
df["Total_Stops"] = df["Total_Stops"].astype(int)
```

```python
df.drop("Route",axis = 1, inplace = True)
```

```python
df
```

```python
df["Duration_h"]=df["Duration"].str.split("h").str[0]
```

```python
df["Duration_m"]=df["Duration"].str.split("h").str[1].str.split("m").str[0]
```

```python
df["Duration_h"]=df["Duration_h"].astype(int)
```

```python
df[df["Duration_h"]=="5m"]
```

```python
df.drop([6474,2660],axis = 0,inplace = True)
```

```python
df["Duration_h"]=df["Duration_h"].astype(int)
```

```python
## replacing '' empty string values with nan
df["Duration_m"]=df.Duration_m.replace('',np.nan,regex = True)
```

```python
df["Duration_m"].fillna(0,inplace = True)
```

```python
df["Duration_m"] = df["Duration_m"].astype(int)
```

```python
df["Duration"] = df["Duration_h"]*60 + df["Duration_m"]
```

```python
df.drop(["Duration_h","Duration_m"], axis = 1,inplace = True)
```
